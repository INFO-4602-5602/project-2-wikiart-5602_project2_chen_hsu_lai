```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('WikiArtClean.csv')
```

```python
# Style vs Mean rating
data1 = data[data['Style'] != 'Modern Art,Post Renaissance Art'] # only one sample
f, ax = plt.subplots(figsize=(20, 6))
sns.set(style="whitegrid")
ax = sns.violinplot(x="Style", y="Mean rating", data=data1, ax=ax)
```

```python
# Category vs Mean Rating
data2 = data
for category in data['Category'].unique():
    if data[data['Category']==category].shape[0] <= 5:
        data2 = data2[data2['Category']!=category]
f, ax = plt.subplots(figsize=(100, 6))
sns.set(style="whitegrid")
ax = sns.violinplot(x="Category", y="Mean rating", data=data2, ax=ax, cut=3)
```

```python
# Century vs Mean Rating
new_data = []
for i, row in data.iterrows():
    new_data.append((str(row['Year']//100+1), row['Mean rating']))
new_data = pd.DataFrame(new_data, columns=['Century', 'Mean rating'])

f, ax = plt.subplots(figsize=(20, 6))
sns.set(style="whitegrid")
ax = sns.violinplot(x="Century", y="Mean rating", data=new_data, ax=ax)
```
